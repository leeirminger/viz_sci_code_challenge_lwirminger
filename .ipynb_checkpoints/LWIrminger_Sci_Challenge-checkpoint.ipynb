```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
inputdir = r"C:\Temp\Demo\Input"
```

```python
#Create geopandas dataframes from the downloaded protected areas shapefiles

protected_areas0_gdf = gpd.read_file(inputdir + r'\WDPA_WDOECM_Mar2022_Public_all_shp\WDPA_WDOECM_Mar2022_Public_all_shp_0\WDPA_WDOECM_Mar2022_Public_all_shp-polygons.shp')
protected_areas1_gdf = gpd.read_file(inputdir + r'\WDPA_WDOECM_Mar2022_Public_all_shp\WDPA_WDOECM_Mar2022_Public_all_shp_1\WDPA_WDOECM_Mar2022_Public_all_shp-polygons.shp')
protected_areas2_gdf = gpd.read_file(inputdir + r'\WDPA_WDOECM_Mar2022_Public_all_shp\WDPA_WDOECM_Mar2022_Public_all_shp_2\WDPA_WDOECM_Mar2022_Public_all_shp-polygons.shp')
```

```python
#Open SPAM Data CSV as pandas dataframe with correct encoding, limited to needed columns
soybean_df = pd.read_csv(inputdir + r'\spam2010v2r0_global_phys_area\spam2010V2r0_global_A_TA.csv', usecols = ['iso3','prod_level','alloc_key','cell5m','x','y','rec_type','tech_type','unit','soyb_a','crea_date','year_data','source','name_cntr','name_adm1','name_adm2'], encoding='cp1252')
```

```python
#Create geopandas gdf from df

soybean_gdf = gpd.GeoDataFrame(
    soybean_df, geometry=gpd.points_from_xy(soybean_df.x, soybean_df.y))
soybean_gdf.crs = 'epsg:4326'
```

```python
#Open bounding boxes needed for challenge

boundingboxes_gdf = gpd.read_file(inputdir + r'\areas.geojson')
```

```python
#Ensure correct CRS for bounding boxes

boundingboxes_gdf.to_crs('epsg:4326')
```

```python
#Get world baselayer, for the sake of interest

world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
```

```python
#Clip the protected  areas
protected_areas0_clipped = gpd.clip(protected_areas0_gdf, boundingboxes_gdf)
protected_areas1_clipped = gpd.clip(protected_areas1_gdf, boundingboxes_gdf)
protected_areas2_clipped = gpd.clip(protected_areas1_gdf, boundingboxes_gdf)

```

```python
#Clip the SPAM yield data


soybeans_clipped = gpd.clip(soybean_gdf, boundingboxes_gdf)
```

```python
# Using the drop() function to remove values of zero in 'soyb_a' field
soybeans_clipped_dropped = soybeans_clipped.loc[~((soybean_df['soyb_a'] == 0))]
```

```python
#Create Map of Soybeans Physical Area and Protected Lands along with the world map for context


ax = soybeans_clipped_dropped.plot(markersize=.01,
                                   figsize=(20, 8),
                                   column='soyb_a',
                                   scheme='quantiles',
                                   k=6,
                                   cmap='jet',
                                   legend=True,
                                   )
                                   
plt.autoscale(False)

plt.style.context(("seaborn", "ggplot"))

world.to_crs(soybeans_clipped_dropped.crs).plot(
    ax=ax,
    color='none',
    edgecolor='grey')

protected_areas0_clipped.to_crs(soybeans_clipped_dropped.crs).plot(
    ax=ax, 
    color='green', 
    edgecolor='green', 
    alpha = .8)
protected_areas1_clipped.to_crs(soybeans_clipped_dropped.crs).plot(
    ax=ax, color='green',
    edgecolor='green',
    alpha = .8)
protected_areas2_clipped.to_crs(soybeans_clipped_dropped.crs).plot(
    ax=ax,
    color='green',
    edgecolor='green',
    alpha = .8)

ax.set(title='Soybean Physical Areas (Hectares) \n and Protected Lands in dark green')
plt.show()

```

```python
#Soybeans, group and sum area

soybeans_clipped_dropped_grouped = soybeans_clipped_dropped.groupby('name_cntr')['soyb_a'].sum()
soybeans_clipped_dropped_grouped
soybeans_clipped_dropped_grouped = pd.DataFrame(soybeans_clipped_dropped_grouped).reset_index()
```

```python
#Convert crs to projected cs for area measurement

protected_areas0_clipped6933 = protected_areas0_clipped.to_crs('epsg:6933')
protected_areas1_clipped6933 = protected_areas1_clipped.to_crs('epsg:6933')
protected_areas2_clipped6933 = protected_areas2_clipped.to_crs('epsg:6933')
```

```python
#Find areas of protected areas


protected_areas0_clipped6933['Area'] = (protected_areas0_clipped6933.area / 10**4)
protected_areas1_clipped6933['Area'] = (protected_areas1_clipped6933.area / 10**4)
protected_areas2_clipped6933['Area'] = (protected_areas2_clipped6933.area / 10**4)

```

```python
#Group by and sum protected areas

protected_areas0_clipped_grouped = protected_areas0_clipped6933.groupby('PARENT_ISO')['Area'].sum()
protected_areas1_clipped_grouped = protected_areas1_clipped6933.groupby('PARENT_ISO')['Area'].sum()
protected_areas2_clipped_grouped = protected_areas2_clipped6933.groupby('PARENT_ISO')['Area'].sum()
```

```python
#Concat and Group Protected Areas 

protected_areasall = pd.concat([protected_areas0_clipped_grouped, protected_areas1_clipped_grouped, protected_areas2_clipped_grouped], sort=False)
protected_areasall_grouped = protected_areasall.groupby('PARENT_ISO').sum()
protected_areasall_grouped = pd.DataFrame(protected_areasall_grouped)
protected_areasall_grouped = protected_areasall_grouped.reset_index()



```

```python
protected_areasall_grouped.plot.scatter(x='PARENT_ISO', y='Area', figsize=(8,6))
plt.show()
```
